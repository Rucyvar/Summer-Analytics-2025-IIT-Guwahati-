```python
import pandas as pd
```

```python
df_test = pd.read_csv("https://raw.githubusercontent.com/Rucyvar/Summer-Analytics-2025-IIT-Guwahati-/refs/heads/main/Week-2%20Assignment/hacktest.csv")
df_train = pd.read_csv("https://raw.githubusercontent.com/Rucyvar/Summer-Analytics-2025-IIT-Guwahati-/refs/heads/main/Week-2%20Assignment/hacktrain.csv")
```

```python
# Impute missing values in numerical columns of df_train with the mean of their respective columns
numerical_cols = df_train.select_dtypes(include=['number']).columns
df_train[numerical_cols] = df_train[numerical_cols].fillna(df_train[numerical_cols].mean())

# Separate the target variable 'class' from the features
X_train = df_train.drop('class', axis=1)
y_train = df_train['class']

# Save the original 'ID' column from df_test
test_ids = df_test['ID'].copy()

# Drop the 'Unnamed: 0' and 'ID' columns from both training features and test DataFrame
# Check if columns exist before dropping to avoid errors
cols_to_drop = ['Unnamed: 0', 'ID']
X_train = X_train.drop(columns=[col for col in cols_to_drop if col in X_train.columns])
df_test = df_test.drop(columns=[col for col in cols_to_drop if col in df_test.columns])


# Encode the categorical target variable 'class' into numerical labels
from sklearn.preprocessing import LabelEncoder, StandardScaler

label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
df_test_scaled = scaler.transform(df_test)

display(X_train_scaled[:5])
display(df_test_scaled[:5])
display(y_train_encoded)
```

```python
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
# Increased max_iter for convergence and added multi_class='auto' for multiclass classification
model = LogisticRegression(max_iter=10000, multi_class='auto', random_state=42)

# Fit the model to the processed training data
model.fit(X_train_scaled, y_train_encoded)
```

```python
y_pred_encoded = model.predict(df_test_scaled)
```

```python
# Decode the predicted numerical labels back to their original class names
y_pred = label_encoder.inverse_transform(y_pred_encoded)

# Create a new pandas DataFrame for the submission file
submission_df = pd.DataFrame({
    'ID': test_ids,
    'class': y_pred
})

# Save the submission DataFrame to a CSV file named submission.csv
submission_df.to_csv('submission.csv', index=False)
```
